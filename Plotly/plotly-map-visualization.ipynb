```python
import numpy as np
import pandas as pd

import plotly
import plotly.plotly as py
plotly.offline.init_notebook_mode(connected=True)
```

# 0. Import Data

```python
countries=pd.read_csv('countries.csv',header=0)
data_2015=pd.read_csv('2015.csv',header=0)
```

```python
data1=data_2015.merge(countries, how='left', left_on='Country',right_on='name')
```

```python
data1.to_csv('data.csv')
```

```python
data1.head()
```

# 1. Choropleth Maps

```python
country_code = pd.read_csv('code.csv',header=0)
country_code.head()
```

```python
data2=data1.merge(country_code,how='left',left_on='country',right_on='Alpha-2 code').drop('ISO 3166-2',axis=1)
data2.to_csv('data.csv')
```

```python
data2
```

```python
data = [ dict(
        type = 'choropleth',
        locations = data2['Alpha-3 code'],
        z = data2['Happiness Score'],
        text = data2['Country'],
#         colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
#             [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = True,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Happiness Score'),
      ) ]

layout = dict(
    title = '2015 World Happiness Score',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```

# 2. Symbol Maps

```python
data2.describe()
```

```python
data3=data2.sort_values('Happiness Score',axis=0,ascending=False)
```

```python
data3['text'] = data3['Country'] + '<br>Happiness Score ' + (data3['Happiness Score']).astype(str)
limits = [(0,2),(3,7),(8,16),(17,50),(51,157)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","lightgrey"]
countries = []

for i in range(len(limits)):
    lim = limits[i]
    df_sub = data3[lim[0]:lim[1]]
    country = dict(
        type = 'scattergeo',
        locationmode = "country names",
        lon = df_sub['longitude'],
        lat = df_sub['latitude'],
        text = df_sub['text'],
        marker = dict(
            size = np.exp(df_sub['Happiness Score'])/5,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    countries.append(country)
```

```python
layout = dict(
        title = '2015 World Happiness Score',
        showlegend = True,
        geo = dict(
            projection=dict( type='world' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict( data=countries, layout=layout )
py.iplot( fig, validate=False, filename='d3-bubble-map-populations' )
```
