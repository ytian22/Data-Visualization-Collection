```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/Users/ty/Downloads/publishers.csv',header=0)
```

```python
data
```

### 1. Bar chart

Among all the genres, "genre fiction" has the highest number of units sold, and foreign language has the lowest.

```python
sns.set(font_scale=1.5)
sns.barplot(x='genre', y='units sold',data=data).set_title('Number of Units Sold by Genre')
plt.show()
```

### 2. Boxplot

Books that belong to children genre has the highest median and first quantile of average ratings, while foreign language has the lowest.

```python
sns.set(font_scale=1.5)
sns.boxplot(x='genre',y='average rating',data=data).set_title('Distribution of Average Rating by Genre')
plt.show()
```

### 3. Scatter plot - bivariate

Understandably, it is more frequent for a relatively higher author revenue happens when there is a higher average rating (>3).

```python
sns.set(font_scale=1.5)
sns.regplot(x='average rating',y='author revenue',data=data,fit_reg=False).set_title('Author Revenue vs Average Rating')
plt.show()
```

### 4. Scatter plot - trivariate

Look from genre perspective, foreign language books and children books do not bring much revenue to authors even with a high average rating, compared to genre fiction, which is the most profitable genre.

```python
sns.lmplot(x='average rating',y='author revenue',data=data,fit_reg=False, hue='genre',size=7).fig.suptitle("Author Revenue vs Average Rating by Genre", fontsize=15)
plt.show()
```

### 5. Swarm plot

Sampled 500 observations from the data, its swarmplot indicates that genre fiction books get more total reviews.

```python
sns.swarmplot(x='genre', y="total reviews", data=data.loc[0:500])
plt.show()
```

### 6. Faceted Histogram

The faceted Histogram of total reviews by genre and type shows an exponential decay trend.

```python
sns.set(style="darkgrid")
g = sns.FacetGrid(data, col="type", row="genre")
bins = np.linspace(0, 100, 20)
g.map(plt.hist, "total reviews", color="steelblue", bins=bins, lw=0)
```

### 7. Extra (violin)

The violin plot of average rating vs genre shows a distribution that foreign language books are more likely to get lower average ratings.

```python
sns.violinplot(x = "genre", y="average rating", data=data)
```
