### Part 1 - Create and visualize a graph 

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data={'E':['A','B','C'],
    'C':['D','A','E'],
    'A':['E','F','G']}
```

```python
data
```

```python
G= nx.from_dict_of_lists(data)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
G.add_node('Z')
G.add_edge('Z','B')
G.add_edge('Z','Y')
nx.draw(G, with_labels=True)
plt.show()
```
